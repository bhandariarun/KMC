```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#y=mx+c
m = int(input('Enter a slope :'))
c = int(input('Enter a intercept :'))
plt.figure(figsize=(10, 5))


x = np.arange(100, 200)

y = m*x+c

plt.plot(x,y, marker='o',color='g')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = mx + c')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```
