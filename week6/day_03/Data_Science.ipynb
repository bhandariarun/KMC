```python
data_path = 'Salary_Data.csv'
df = pd.read_csv(data_path)
df
```

```python
# Read the data into a DataFrame
# data = {
#     'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9],
#     'Salary': [39343.0, 46205.0, 37731.0, 43525.0, 39891.0, 56642.0]
# }
# df = pd.DataFrame(data)

# Define the independent variable (x) and dependent variable (y)
x = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary'].values

# Perform linear regression
model = LinearRegression()
model.fit(x, y)

# Get the slope (coefficient) and intercept of the regression line
slope = model.coef_[0]
intercept = model.intercept_
# slope = 2
# intercept = 3

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(x, slope * x + intercept, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# Print the equation of the regression line
print(f'Equation of the regression line: y = {slope:.2f}x + {intercept:.2f}')
```

```python
#Simple Linear Regression 3*x1+2*x2+4*x3
# Generate x1, x2, and x3 values ranging from 100 to 200
x1 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)
x2 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)
x3 = np.linspace(100, 200, num=len(df)).reshape(-1, 1)

# Calculate y values based on the equation y = 3x1 + 2x2 + 4x3 + 5
y = 3 * x1 + 2 * x2 + 4 * x3 + 5

# Perform simple linear regression for x1, x2, x3 and y
model_x1 = LinearRegression()
model_x1.fit(x1, y)

model_x2 = LinearRegression()
model_x2.fit(x2, y)

model_x3 = LinearRegression()
model_x3.fit(x3, y)

# Plot the data points
plt.scatter(x1, y, color='blue', label='x1')
plt.scatter(x2, y, color='green', label='x2')
plt.scatter(x3, y, color='orange', label='x3')

# Plot the regression lines
plt.plot(x1, model_x1.predict(x1), color='red', label='Regression Line for x1')
plt.plot(x2, model_x2.predict(x2), color='purple', label='Regression Line for x2')
plt.plot(x3, model_x3.predict(x3), color='brown', label='Regression Line for x3')

# Add labels and legend
plt.xlabel('Independent Variables')
plt.ylabel('Dependent Variable (y)')
plt.title('Simple Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
#Multiple Linear Regression and plot the equation
X = df[['YearsExperience']]
y = df['Salary']

# Create and fit the multiple linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate x1, x2, and x3 values ranging from 100 to 200
x1_values = np.linspace(100, 200, num=100).reshape(-1, 1)
x2_values = np.linspace(100, 200, num=100).reshape(-1, 1)
x3_values = np.linspace(100, 200, num=100).reshape(-1, 1)

# Predict the corresponding y values using the trained model
y_predicted = (3 * x1_values) + (2 * x2_values) + (4 * x3_values) + 5

# Plot the predicted values
plt.plot(x1_values, y_predicted, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('x1, x2, x3')
plt.ylabel('Predicted Salary')
plt.title('Multiple Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
x_values = np.linspace(100, 200, num=100).reshape(-1, 1)

# Calculate y values based on the equation y = 2x^2 + 3x + 4
x_squared = x_values ** 2
x_cubed = x_values ** 3
y_values = 2 * x_squared + 3 * x_values + 4

# Perform polynomial regression
poly_features = PolynomialFeatures(degree=2)  # Use degree=2 for quadratic regression
X_poly = poly_features.fit_transform(df[['YearsExperience']])
model = LinearRegression()
model.fit(X_poly, df['Salary'])

# Transform x_values for plotting
x_values_poly = poly_features.transform(x_values)

# Predict y values using the model
y_pred = model.predict(x_values_poly)

# Plot the data points
plt.scatter(df['YearsExperience'], df['Salary'], color='blue', label='Data Points')

# Plot the regression line
plt.plot(x_values, y_pred, color='red', label='Polynomial Regression')

# Add labels and legend
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Polynomial Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# Print the equation of the regression line
print("Equation of the regression line:")
print("y =", model.coef_[2], "* x^2 +", model.coef_[1], "* x +", model.intercept_)
```
