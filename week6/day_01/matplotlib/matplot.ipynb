```python
#1. `How does the average satisfaction level change with increasing years of employment? [Line]
df_hr = df.groupby('time_spend_company').agg({'satisfaction_level':'mean'})
df_hr
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_hr, marker='o', color='g')
plt.title('Average Satisfaction Vs. Years of Employment')
plt.xlabel('Years of Employment')
plt.ylabel('Average Satisfaction')
plt.grid(True)
plt.show()
```

```python
#1. `How does the average satisfaction level change with increasing years of employment? [scatter]
average_satisfaction = df.groupby('time_spend_company')['satisfaction_level'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(average_satisfaction['time_spend_company'], 
            average_satisfaction['satisfaction_level'], 
            color='blue', 
            alpha=0.6)

# Add labels and title
plt.title('Average Satisfaction Level vs. Years of Employment')
plt.xlabel('Years of Employment')
plt.ylabel('Average Satisfaction Level')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#6. `How does employee satisfaction vary across different departments?` [Box Plot]
plt.figure(figsize=(10, 6))
dept_satisfaction = [df[df['Department'] == dept]['satisfaction_level'] for dept in df['Department'].unique()]
plt.boxplot(dept_satisfaction, labels=df['Department'].unique(), vert=False)
plt.title('Employee Satisfaction Across Different Departments')
plt.ylabel('Department')
plt.xlabel('Satisfaction Level')
plt.xticks(rotation=45)  # Rotate department names for better readability if needed
plt.tight_layout()

# Show plot
plt.show()
```

```python
#7. `Figure out employee leaving status (time_spend vs satisfaction_level vs left?`[Bubble Plot]
plt.figure(figsize=(10, 6))

# Define bubble plot
plt.scatter(df['time_spend_company'], 
            df['satisfaction_level'], 
            s=df['left']*100, 
            c=df['left'], 
            cmap='coolwarm', 
            alpha=0.6,
            edgecolors='w',
            linewidth=1)

# Add labels and title
plt.title('Employee Leaving Status based on Time Spent and Satisfaction Level')
plt.xlabel('Time Spent in Company')
plt.ylabel('Satisfaction Level')

# Add a legend
plt.legend(['Not Left', 'Left'], loc='upper right')

# Show plot
plt.grid(True)
plt.colorbar(label='Left')
plt.tight_layout()
plt.show()
```

```python
#8.`How are the numerical variables correlated with each other?`
numerical_data = df[['satisfaction_level', 'last_evaluation', 'number_project', 
                       'average_montly_hours', 'time_spend_company']]

# Calculate correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
#9. `How to create different scatter plots under single matplotlib figure?`
# Sample data
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [3, 5, 7, 11, 13]

x3 = [1, 2, 3, 4, 5]
y3 = [1, 4, 6, 8, 10]

# Create a new figure
plt.figure(figsize=(10, 6))

# Scatter plot 1
plt.scatter(x1, y1, color='red', label='Plot 1')

# Scatter plot 2
plt.scatter(x2, y2, color='blue', label='Plot 2')

# Scatter plot 3
plt.scatter(x3, y3, color='green', label='Plot 3')

# Add legend
plt.legend()

# Add labels and title
plt.title('Multiple Scatter Plots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
