## Questions

- **`1.How does the average satisfaction level change with increasing years of employment?`**    

    - This questions can be answered via:
        - `1.1 Line plot:`
            - Line plots are best suited for trend-based visualizations of data over a period of time.  
            - Line plots are used when dataset is continuous. 

        - `1.2 Scatter plot:`
            - A scatter plot is a useful method of comparing variables against each other.
            - Working is similar to Line plots.
            - In Line plots, data points are connected whereas in scatter plot data points are not connected

`1.1 Line plot`

```python
# calculate average satisfaction level for each years of employment

avg_satisfaction_by_years = df_hr.groupby('time_spend_company').agg({'satisfaction_level': 'mean'})

avg_satisfaction_by_years   
```

```python
# create a line plot of average satisfaction level by years of employment

# define figure size of (10, 6)
plt.figure(figsize=(6, 4))

# extract x_data and y_data of line plot
x_data = avg_satisfaction_by_years.index
y_data = avg_satisfaction_by_years.values

# create line plot
plt.plot(x_data, y_data, marker='o', color='b')

# define x_label, y_label, title
plt.xlabel('Years of Employment')
plt.ylabel('Average Satisfaction Level')
plt.title('Average Satisfaction Level by Years of Employment')
plt.grid(True) # for grid to be displayed
plt.show()
```

- From above,
    - Decrease Trend: At start satisfaction_level is high and it decreases upto 4 years.
    - Increase Trend: After 4 years the satisfaction_level starts increasing.

`1.2 Scatter plot`  


```python
# create a scatter plot of average satisfaction level by years of employment

# define figure size of (10, 6)
plt.figure(figsize=(6, 4))

plt.scatter(x_data, y_data, marker='o', color='red')

# define x_label, y_label, title
plt.xlabel('Years of Employment')
plt.ylabel('Average Satisfaction Level')
plt.title('Average Satisfaction Level by Years of Employment')
plt.grid(True) # for grid to be displayed
plt.show()
```

- This shows scatter plot is quite similar to line plot except dots are not connected by line.

**`Q.4. What is the distribution of work accidents among departments?`**
- This can be answered via: `pie_chart`
    - A pie chart is a circular statistical graphic divided into slices to illustrate numerical propotions.
    - It can be bit more confusing to read compared to bar chart.

```python
# Create a pie chart of work accidents distribution by department

accident_counts = df_hr.groupby('Department')['Work_accident'].sum()

print(accident_counts)
```

```python
# define figure size
plt.figure(figsize=(6, 4))

# create pie chart
pie_labels = accident_counts.index
pie_data = accident_counts.values

# define how much to explode different pie wedges
# 0 indicates do not explode
explode = [0, 0, 0, 0, 0, 0, 0, 0.09, 0, 0]

plt.pie(
    pie_data, 
    labels=pie_labels, 
    autopct='%1.1f%%', 
    startangle=140, 
    shadow=True,
    explode=explode,
    colors=plt.cm.Paired.colors)

plt.title('Work Accident Distribution by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

- This shows that `Sales` department has highest number of work accident of 27.1%

**`Q.7 Figure out employee leaving status (time_spend vs satisfaction_level with left as HUE)?`**
- This can be answered via: `Bubble Plot`
    - A bubble plots is a variation of the scatter plot that displays three dimensions of data (x, y, z).
    - The data points are replace with bubbles.
    - The size of the Bubbles is determined by the third variable `z` also know as the weight.

```python
stats_by_time_spend = df_hr.groupby('time_spend_company').agg({'satisfaction_level': 'mean', 
                                                               'left': 'sum'}).reset_index()
x_data = stats_by_time_spend['time_spend_company']
y_data = stats_by_time_spend['satisfaction_level']
z_data = stats_by_time_spend['left']

# normalize weights i.e. z_data to be in the range of 0 to 1
z_weights = (z_data - z_data.min()) / (z_data.max() - z_data.min())
```

```python
# create a bubble plot
plt.figure(figsize=(10, 6))

# multiply normaized_z by 500 to scale it on graph
# add 20 to normalized_z to compensate for minimum value which has 0 weight.

ax = plt.scatter(x_data, y_data, alpha=0.8, c=z_weights, cmap='coolwarm', s=z_weights*500 + 20) # c represents number denoted by colors (using z data)

# set title, xlabel, ylabel
plt.xlabel('Time Spend (Years)', fontweight='bold')
plt.ylabel('Satisfaction Scores', fontweight='bold')
plt.title('Employee Leaving Status', fontsize=16, fontweight='bold')

# set color bar
color_bar = plt.colorbar(ax)
color_bar.set_label('Left', fontweight='bold')

plt.grid(linestyle='--')
plt.show()
```

- Upto 4 Years, employee satisfaction_level decreases (least satisfaction_level at 4 years) with increase in employee churn.
- After 4 Years, employee satisfaction_level started inreasing and employee churn started decreasing.
- This shows that, Company should focus on develop strategy to retain employees with time spend in a company less than or equal to 4 years.

**`Q.8 How are the numerical variables correlated with each other?`**
- This can be answered via: `Heatmap`
    - A Heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours.
    - It is difficult to interpret numbers in a matrix, thus comes Heatmap.

- In matplotlib, there is no predefined function to draw heatmap, However we can use `plt.imshow()` function to draw heatmap.

```python
# get correlation matrix
# Hint: df.corr()

# correlation_matrix = df_hr.corr()
df_hr_numeric = df_hr.select_dtypes(include=['float64', 'int64'])
correlation_matrix = df_hr_numeric.corr()

# correlation_matrix
```

```python
correlation_matrix.columns
```

```python
# Create Heatmap of Correlation Matrix
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')

# Add annotations
# using nested loops and text() function from matplotlib
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        # plt.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='white' if correlation_matrix.iloc[i, j] < 0.5 else 'black')
        plt.text(j, i, 
                f"{correlation_matrix.iloc[i, j]:.2f}", 
                ha='center', 
                va='center',
                color='white' if correlation_matrix.iloc[i, j] < 0.5 else 'black')


# define colorbar to interpret color range
plt.colorbar(heatmap)

# add xticks and yticks
# to redefine xticks label and yticks label with column names
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.show()
```

- From the above Heatmap,
    - Correlation in diagonal is 1 (high) since correlation of variables with the same variables is always high.
    - Corrrelation between satisfaction_level and left column is `-0.39` (negatively correlated)
        - This means, increase in satisfaction_level decreases the chance of employee leaving company.

**`Q. 9 How to create different scatter plots under single matplotlib figure?`**
- This can be answered via `subplots` concept.  

    - subplots method provids a way to plot multiple plots on a single figure.
    - Subplots are particularly useful when you want to show different views or relationships between variables in a compact and organized manner.
    - Given number of rows and columns of figure, it returns figure axis location.

- Let's see how we can create a subplots of scatter plot.

```python
# Create subplots with adjusting space of scatter plots

fig, axes = plt.subplots(3, 2, figsize=(6, 10))

# create subset dataframe for left = 1 and left = 0
left_df = df_hr[df_hr['left'] == 1]
not_left_df = df_hr[df_hr['left'] == 0]

# x_axis_label_list = ['satisfaction_level', 'last_evaluation']
# y_axis_label_list = []

variables_dict = {
    'satisfaction_level': ['last_evaluation', 'number_of_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years'],
    'last_evaluation': ['satisfaction_level', 'number_of_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']
}


# define left columns  colors
colors = {0: 'blue', 1: 'orange'}

# SUBPLOTS 1 (0, 0)
# satisfaction_level vs last_evaluatio
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[0, 0].scatter(subset['satisfaction_level'], subset['last_evaluation'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')
    axes[0, 0].set_ylabel('last_evaluation')
# SUBPLOTS 2 (1, 0)
# satisfaction_level vs average_montly_hours
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[1, 0].scatter(subset['satisfaction_level'], subset['average_montly_hours'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')
    axes[1, 0].set_ylabel('average_montly_hours')

# SUBPLOTS 3 (2, 0)
# satisfaction_level vs number_project
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[2, 0].scatter(subset['satisfaction_level'], subset['number_project'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')
    
    axes[2, 0].set_xlabel('satisfaction_level')
    axes[2, 0].set_ylabel('number_project')

# SUBPLOTS 4 (0, 1)
# last_evaluation vs satisfaction_level
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[0, 1].scatter(subset['last_evaluation'], subset['satisfaction_level'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')

    axes[0, 1].set_ylabel('satisfaction_level')
# SUBPLOTS 5 (1, 1)
# last_evaluation vs average_montly_hours
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[1, 1].scatter(subset['last_evaluation'], subset['average_montly_hours'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')

    axes[1, 1].set_ylabel('average_montly_hours')
# SUBPLOTS 6 (2, 1)
# last_evaluation vs number_project
for label, color in colors.items():
    subset = df_hr[df_hr['left'] == label]
    axes[2, 1].scatter(subset['last_evaluation'], subset['number_project'], alpha=0.5, edgecolors='white', marker='s' if label == 1 else 'o')

    axes[2, 1].set_ylabel('number_project')
    axes[2, 1].set_xlabel('last_evaluation')

# Set a single title for the entire set of subplots
fig.suptitle('HR Analytics Subplots', fontsize=16, fontweight='bold')

# Adjust height and width space between subplots
plt.subplots_adjust(hspace=0.3, wspace=0.3) 

# SUBPLOTS 
plt.show()
```
