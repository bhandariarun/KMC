```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.worldometers.info/coronavirus/'
x = requests.get(url)
print(x)
```

```python
soup = BeautifulSoup(x.content, 'html.parser')
print(soup.prettify())[:5000]
```

```python
y = soup.find('table', id='main_table_countries_today')
print(y)
```

```python
z = soup.find('tbody').find_all('tr')
print(z)
```

```python
for data in z:
    print(data)
    break
```

```python
column_names = [
    "Names",
    "Total Cases",
    "New Cases",
    "Total Deaths",
    "New Deaths",
    "Total Recovered",
    "New Recovered",
    "Active Cases",
    "Serious Cases",
    "Total tests",
    "Population"
]
```

```python
complete_data = []
for i in range(8, len(z)):
    data = []
    list_data = z[i].find_all('td')
    for i in list_data:
        data.append(i.text)
    complete_data.append(data)
print(complete_data)
```

```python
headers=[i for i in column_names]
data=[j for j in complete_data[:8]]
output_data = [headers]
print(output_data)
print(data)
```

```python
import pandas as pd
```

```python
for i in range(0, len(data), len(headers)):
        row = data[i:i + len(headers)]
        output_data.append(row)

# for row in output_data:
#     print(row)
df=pd.DataFrame(output_data)
print(df)
```

```python
import csv
csv_file_path = 'output.csv'
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(output_data)
print(f'Data has been saved to {csv_file_path}')
df.to_csv("output.csv", index=False)
```
